```python
# We can see that the rich people in class 1 are of older age than the second and third class.

# so we shall replace the average age in place of null values
```

```python
# function to replace the age in null values

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 24
    else:
        return Age
```

```python
# applying this function to replace null values

train["Age"] = train[["Age", "Pclass"]].apply(impute_age, axis = 1)
```

```python
sns.heatmap(train.isnull())
```

```python
# WE can see that the no null values are there in age column

# again we see that the cabin col has more null values.
# we shall now drop this column
```

```python
train.drop("Cabin", axis=1, inplace=True)
```

```python
pd.get_dummies(train["Sex"], drop_first=True).head(10)
```

```python
train.drop(61, axis=0, inplace=True)
```

```python
# we can not work on data with category
# so we will change them to numerical category

sex = pd.get_dummies(train["Sex"], drop_first=True)
embark = pd.get_dummies(train["Embarked"], drop_first=True)
```

```python
# drop the unwanted columns

train.drop(["Sex", "Embarked", "Name", "Ticket"], axis = 1, inplace = True)
```

```python
train.head()
```

```python
train.columns
```

```python
train = pd.concat([train,sex,embark], axis = 1)
```

```python
train.head()
```

```python
# Train and test data split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train.drop("Survived",axis = 1),
                                                   train["Survived"], test_size = 0.30,
                                                    random_state = 101)
```

```python
#print(x_train)
#print(x_test)
#print(y_train)
#print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test, predictions)
accuracy
```
